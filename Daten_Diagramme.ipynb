```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
plt.style.use({'figure.facecolor':'white'})
```

```python
sheetnames=[]
with pd.ExcelFile('data/2021-04-22_Daten Diagramme.xlsx') as sheets:
    sheetnames=sheets.sheet_names

display(sheets.sheet_names)
sheetdata={sheetname:pd.read_excel('data/2021-04-22_Daten Diagramme.xlsx',sheet_name=sheetname,header=[0]) for sheetname in sheets.sheet_names}
```

```python
plt.style.use('C:\\Users\\maede\\.matplotlib\\diss_martin_arh.mplstyle')
display(sheetdata['Schmelzextraktion SZ'])
fig=plt.figure()
ax=fig.add_axes([0.25,0.166666667,0.49,0.66666667])
rage=np.arange(0,3,1)
ax.bar(-0.4,sheetdata['Schmelzextraktion SZ'].iloc[3,0+1],width=0.5,zorder=1.5,color='lightgrey',yerr=sheetdata['Schmelzextraktion SZ'].iloc[4,0+1],capsize=5,
                 alpha=1,edgecolor='black',lw=1,label=sheetdata['Schmelzextraktion SZ'].columns[0+1])
ax.bar(1,sheetdata['Schmelzextraktion SZ'].iloc[3,1+1],width=0.5,zorder=1.5,color='lightgrey',yerr=sheetdata['Schmelzextraktion SZ'].iloc[4,1+1],capsize=5,
                 alpha=1,edgecolor='black',lw=1,label=sheetdata['Schmelzextraktion SZ'].columns[1+1])
ax.bar(2.4,sheetdata['Schmelzextraktion SZ'].iloc[3,2+1],width=0.5,zorder=1.5,color='lightgrey',yerr=sheetdata['Schmelzextraktion SZ'].iloc[4,2+1],capsize=5,
                 alpha=1,edgecolor='black',lw=1,label=sheetdata['Schmelzextraktion SZ'].columns[2+1])
    #ax.set_xticklabels({i:sheetdata['Schmelzextraktion SZ'].columns[i+1]})
ax.set_xticks([-0.4,1,2.4])
ax.tick_params(axis='x',which='both',length=0)
ax.set_ylabel('Wasserstoffgehalt [ppm]')
ax.set_xlim([-1,3])
ax.set_xticklabels(sheetdata['Schmelzextraktion SZ'].columns[1:],rotation=0)
ax.set_xlabel('Schweißzusatz')
fig.savefig('Schmelzextraktion.svg',dpi=500)
```

```python
display(sheetdata['Pulverfeuchte'])
fig=plt.figure()
ax=fig.add_axes([0.20,0.16666667,0.49,0.66666667])
rage=np.arange(0,2,1)

ax.bar(2,sheetdata['Pulverfeuchte'].iloc[0,0+1],width=0.5,zorder=1.5,
           color='lightgrey',yerr=sheetdata['Pulverfeuchte'].iloc[1,0+1],capsize=5,alpha=1,edgecolor='black',lw=1)
ax.bar(0,sheetdata['Pulverfeuchte'].iloc[0,1+1],width=0.5,zorder=1.5,
           color='lightgrey',yerr=sheetdata['Pulverfeuchte'].iloc[1,1+1],capsize=5,alpha=1,edgecolor='black',lw=1)
ax.set_xticks([0,2])
ax.tick_params(axis='x',which='both',length=0)
ax.set_xlim([-1,3])
ax.set_ylabel('rel. Pulverfeuchte [%]')# H$_{2}$O
ax.set_xticklabels(['trocken','feucht'],rotation=0)
ax.set_xlabel('Pulverzustand')
fig.savefig('Pulverfeuchte.svg',dpi=500)
```

```python
display(sheetdata['H-Gehalte Nitrogen'])
fig=plt.figure()
ax=fig.add_axes([0.125,0.16666667,0.75,0.66666667])
display(len(sheetdata['H-Gehalte Nitrogen']))
rage=range(0,len(sheetdata['H-Gehalte Nitrogen']),2)
ax.tick_params(axis='x',which='both',length=0)
for i in rage:
    ax.bar(i/2,sheetdata['H-Gehalte Nitrogen'].iloc[i,3],width=0.5,zorder=1.5,
           color='lightgrey',yerr=sheetdata['H-Gehalte Nitrogen'].iloc[i+1,3],capsize=5,alpha=1,edgecolor='black',lw=1)
    ax.bar(i/2,sheetdata['H-Gehalte Nitrogen'].iloc[i,5],width=0.5,zorder=1.0,
           color='grey',yerr=sheetdata['H-Gehalte Nitrogen'].iloc[i+1,5],capsize=5,alpha=1,edgecolor='black',lw=1)
ax.set_xticks([0,1,2,3,4,5])
ax.set_xticklabels(sheetdata['H-Gehalte Nitrogen'].loc[::2,'label'])
ax.text(0.5,1.7,'S355N-S2',ha='center',va='center')
ax.text(2.5,1.7,'L555M-\nS3Ni1Mo',ha='center',va='center')
ax.text(4.5,1.7,'S690QL-\nS3Ni2,5CrMo',ha='center',va='center')

ax.set_ylabel('Wasserstoffgehalt [ppm]')
ax.set_xlabel('Pulverzustand')
ax.axvline(1.5,color='black')
ax.axvline(3.5,color='black')
ax.set_xlim([-0.5,5.5])
ax.set_ylim([0,2])
fig.savefig('H-Gehalte Nitrogen.svg',dpi=500)
```
