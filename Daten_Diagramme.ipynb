```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
plt.style.use({'figure.facecolor':'white'})
```

```python
sheetnames=[]
with pd.ExcelFile('data/2021-04-22_Daten Diagramme.xlsx') as sheets:
    sheetnames=sheets.sheet_names

display(sheets.sheet_names)
sheetdata={sheetname:pd.read_excel('data/2021-04-22_Daten Diagramme.xlsx',sheet_name=sheetname,header=[0]) for sheetname in sheets.sheet_names}
```

```python
plt.style.use('C:\\Users\\maede\\.matplotlib\\diss_martin_general.mplstyle')
display(sheetdata['Schmelzextraktion SZ'])
fig=plt.figure()
ax=fig.add_axes([0.132,0.125,0.75,0.75])
rage=np.arange(0,3,1)
for i in rage:
    ax.bar(i,sheetdata['Schmelzextraktion SZ'].iloc[3,i+1],width=0.5,zorder=1.5,color='blue',yerr=sheetdata['Schmelzextraktion SZ'].iloc[4,i+1],capsize=5,
                 alpha=1,edgecolor='grey',lw=1,label=sheetdata['Schmelzextraktion SZ'].columns[i+1])
    #ax.set_xticklabels({i:sheetdata['Schmelzextraktion SZ'].columns[i+1]})
ax.set_xticks(list(rage))
ax.set_xticklabels(sheetdata['Schmelzextraktion SZ'].columns[1:],rotation=0)
```

```python
display(sheetdata['Pulverfeuchte'])
fig=plt.figure()
ax=fig.add_axes([0.132,0.125,0.75,0.75])
rage=np.arange(0,2,1)
for i in rage:
    ax.bar(i,sheetdata['Pulverfeuchte'].iloc[0,i+1],width=0.5,zorder=1.5,color='blue',yerr=sheetdata['Pulverfeuchte'].iloc[1,i+1],capsize=5,
                 alpha=1,edgecolor='grey',lw=1)
ax.set_xticks(list(rage))
ax.set_xticklabels(sheetdata['Pulverfeuchte'].columns[1:],rotation=0)
```
